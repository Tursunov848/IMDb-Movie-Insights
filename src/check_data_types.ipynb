```python
import pandas as pd

df = pd.read_csv("top_250_enriched.csv")

df["BoxOffice_Million"] = df["BoxOffice"] / 1_000_000
```

```python
def format_box_office(x):
    if pd.isna(x):
        return "N/A"
    if x >= 1_000_000_000:
        return f"${x/1_000_000_000:.1f}B"
    elif x >= 1_000_000:
        return f"${x/1_000_000:.1f}M"
    elif x >= 1_000:
        return f"${x/1_000:.1f}K"
    else:
        return f"${x:.0f}"

df["BoxOffice_Label"] = df["BoxOffice"].apply(format_box_office)
```

```python
df["BoxOffice_Missing"] = df["BoxOffice"].isna()
```
